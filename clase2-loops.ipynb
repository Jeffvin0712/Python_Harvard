# While

`while` es una estructura de control que repite un bloque de código **mientras** una condición sea verdadera.  
Cuando la condición deja de cumplirse, el ciclo termina.

**Ejemplo:**

```python
x = 0

while x < 5:
    print(x)
    x += 1
# Output:
# 0
# 1
# 2
# 3
# 4
```

En este ejemplo, el ciclo `while` se ejecuta mientras `x` sea menor que 5. En cada iteración, se imprime el valor de `x` y luego se incrementa en 1. Cuando `x` alcanza el valor de 5, la condición `x < 5` deja de ser verdadera, y el ciclo termina.

````python
i = 3

while i > 0:
    print("Meow")
    i -= 1
````

````python
i = 3
while i != 0:
    print("Meow")
    i = i - 1
````

````python
i = 1
while i <= 3:
    print("Meow")
    i = i + 1
````

# For

`for` es una estructura de control que repite un bloque de código **iterando** sobre una secuencia (lista, cadena, rango, etc.).  
Se usa cuando sabemos cuántas veces queremos repetir o recorrer elementos.

**Ejemplo:**

```python
for i in range(5):
    print(i)
# Output:
# 0
# 1
# 2
# 3
# 4
```
En este ejemplo, el ciclo `for` itera sobre los números generados por `range(5)`, que produce los números del 0 al 4. En cada iteración, se imprime el valor de `i`. Cuando se han recorrido todos los números, el ciclo termina.

````python
for i in [1, 2, 3]: # esta es una forma de iterar sobre una lista de números con un rango de números manual, pero no es la forma más común de hacerlo en Python.
    print("Meow")
````

````python
for i in range(3): # range(n) genera una secuencia de números desde 0 hasta n-1 y n-1 es el número de iteraciones que se van a realizar
    print("Meow")

for _ in range(3): # la linea _ es un variable de uso común para indicar que no se va a usar el valor de la variable, es decir, que no se va a usar el número de iteraciones.
    print("_Meow")
````

````python
print("Meow" * 3) # esta es una forma de imprimir una cadena de texto un número determinado de veces, en este caso, 3 veces.
print("Meow\n" * 3) #el caracter \n es un salto de línea, por lo que esta línea de código imprimirá "Meow" 3 veces, cada una en una línea diferente.

print("Meow\n" * 3, end="") # el argumento end="" es para evitar que se imprima un salto de línea al final de la última línea, por lo que esta línea de código imprimirá "Meow" 3 veces, cada una en una línea diferente, pero sin un salto de línea al final.
````

````python
while True:
    n = int(input("¿Cuántas veces quieres que se imprima 'Meow'? "))
    if n < 0:
        continue # hace el programa continue con la siguiente iteración del bucle, es decir, vuelve a pedir al usuario que ingrese un número.
    else:
        break # Rompe el bucle, es decir, sale del bucle y continúa con el resto del programa.

for _ in range(n):
    print("Meow")
````

# Comparación: For vs While

| Característica        | For                                   | While                                |
|------------------------|---------------------------------------|--------------------------------------|
| Uso principal         | Iterar sobre una secuencia conocida   | Repetir mientras una condición sea verdadera |
| Control del ciclo     | Basado en elementos de lista, rango, etc. | Basado en condición booleana          |
| Número de iteraciones | Generalmente conocido de antemano     | Puede ser desconocido o indefinido    |
| Comodidad             | Ideal para recorrer colecciones       | Ideal para bucles dependientes de estado |
| Ejemplo               | `for i in range(5):`                 | `while x < 5:`                        |

En resumen, `for` es más adecuado para iterar sobre secuencias o cuando el número de iteraciones es conocido, mientras que `while` es más flexible para situaciones donde la repetición depende de una condición que puede cambiar durante la ejecución del ciclo.

# List

Una `list` es una estructura de datos en Python que permite almacenar **múltiples valores** en una sola variable.  
Los elementos pueden ser de distintos tipos (números, cadenas, booleanos, etc.) y se acceden por su **índice**.

**Características:**
- Ordenada (mantiene el orden de inserción).
- Mutable (se puede modificar después de creada).
- Indexada desde `0`.

**Ejemplo:**

```python
frutas = ["manzana", "banana", "cereza"]

print(frutas[0])   # "manzana"
print(frutas[1])   # "banana"

frutas.append("naranja")   # agrega un elemento
print(frutas)  # ["manzana", "banana", "cereza", "naranja"]
frutas[1] = "kiwi"  # modifica un elemento
print(frutas)  # ["manzana", "kiwi", "cereza", "naranja"]
``` 
En este ejemplo, se crea una lista llamada `frutas` con tres elementos. Se accede a los elementos usando índices, se agrega un nuevo elemento con `append()`, y se modifica un elemento asignando un nuevo valor a su índice.

````python
students = ["Hermione", "Harry", "Ron"]

print(students[0]) # imprime el primer elemento de la lista, que es "Hermione"
print(students[1]) # imprime el segundo elemento de la lista, que es "Harry
print(students[2]) # imprime el tercer elemento de la lista, que es "Ron"
````

````python
students = ["Hermione", "Harry", "Ron"]

for student in students:
    print(student) # imprime cada elemento de la lista, es decir, "Hermione", "Harry" y "Ron"
````

````python
students = ["Hermione", "Harry", "Ron"]

for i in range(students): # esta línea de código no es correcta, ya que range() espera un número como argumento, pero students es una lista, por lo que se generará un error.
    print(i + 1, students[i])
````

````python
students = ["Hermione", "Harry", "Ron"]

for i in range(len(students)): # len(students) devuelve el número de elementos en la lista, que es 3 en este caso, por lo que el rango de números será de 0 a 2, es decir, 0, 1 y 2.
    print(i + 1, students[i]) # imprime el número de iteración (i + 1) y el elemento de la lista correspondiente a esa iteración (students[i]), es decir, "1 Hermione", "2 Harry" y "3 Ron".
````

# Dictionary

Un `dictionary` es una estructura de datos en Python que almacena **pares clave-valor**.  
Cada elemento tiene una **clave única** que sirve para acceder a su valor asociado.

**Características:**
- No está ordenado por defecto (aunque mantiene orden de inserción desde Python 3.7+).
- Mutable (se puede modificar después de creado).
- Las claves deben ser únicas e inmutables (ej. cadenas, números, tuplas).

**Ejemplo:**

```python
persona = {
    "nombre": "Alice",
    "edad": 25,
    "ciudad": "Madrid"
}

print(persona["nombre"])   # "Alice"
print(persona["edad"])     # 25

persona["edad"] = 26       # modificar valor
persona["profesion"] = "Ingeniera"  # agregar nuevo par

print(persona)
# {"nombre": "Alice", "edad": 26, "ciudad": "Madrid", "profesion": "Ingeniera"}
```

En este ejemplo, se crea un diccionario llamado `persona` con tres pares clave-valor iniciales. Se accede a los valores usando sus claves, se modifica el valor de la clave "edad", y se agrega un nuevo par clave-valor para "profesion".

# Diccionario dentro de una List

En Python, una `list` puede contener cualquier tipo de dato, incluyendo **diccionarios**.  
Esto permite tener colecciones de objetos más complejos, donde cada elemento de la lista es un diccionario con pares clave-valor.

**Ejemplo:**

```python
personas = [
    {"nombre": "Alice", "edad": 25},
    {"nombre": "Bob", "edad": 30},
    {"nombre": "Carlos", "edad": 22}
]

# Acceder al primer diccionario
print(personas[0])          # {"nombre": "Alice", "edad": 25}

# Acceder a un valor dentro del diccionario
print(personas[1]["nombre"])  # "Bob"

# Agregar un nuevo diccionario a la lista
personas.append({"nombre": "Diana", "edad": 28})

print(personas)
# [{"nombre": "Alice", "edad": 25}, {"nombre": "Bob", "edad": 30}, {"nombre": "Carlos", "edad": 22}, {"nombre": "Diana", "edad": 28}]
``` 

En este ejemplo, se crea una lista llamada `personas` que contiene tres diccionarios, cada uno representando a una persona con su nombre y edad. Se accede a los diccionarios y a sus valores usando índices y claves, y se agrega un nuevo diccionario a la lista con `append()`.

````python
# los Keys son "name", "house" y "patronus", y los values son los datos correspondientes a cada estudiante, es decir, el nombre del estudiante, la casa a la que pertenece y su patronus.
students = [
    {"name": "Hermione", "house": "Gryffindor", "patronus": "Otter"},
    {"name": "Harry", "house": "Gryffindor", "patronus": "Stag"},
    {"name": "Ron", "house": "Gryffindor", "patronus": "Jack Russell Terrier"},
    {"name": "Draco", "house": "Slytherin", "patronus": None} # el None es un valor especial en Python que representa la ausencia de un valor, es decir, que Draco no tiene un patronus.
]

for student in students:
    print(student["name"], student["house"], student["patronus"], sep=" - ") 
    # imprime el nombre del estudiante (student["name"]), la casa a la que pertenece (student["house"]) y su patronus (student["patronus"])
````

## Dictionary `.get()`

El método `.get()` se utiliza para **obtener el valor asociado a una clave** en un diccionario.  
La diferencia con acceder directamente (`dic["clave"]`) es que `.get()` **no genera error** si la clave no existe, sino que devuelve `None` o un **valor predeterminado** que podemos especificar.

**Sintaxis:**
```python
diccionario.get(clave, valor_predeterminado)
```
- **clave**: la clave que queremos buscar.
- **valor_predeterminado (opcional)**: lo que se devuelve si la clave no existe.
Si no se indica, devuelve None.


````python
def main():
    spacecraft = {"name": "James Webb Space Telescope"}
    print(create_report(spacecraft))
    

def create_report(spacecraft):
    return f"""
    ========= REPORT =========

    Name: {spacecraft.get('name', "Unknown")}
    Distance: {spacecraft.get('distance', "Unknown")} AU

    =========================
    """

main()
````

## Dictionary `.update()`

El método `.update()` se utiliza para **modificar o agregar pares clave-valor** en un diccionario.  
Si la clave ya existe, su valor se sobrescribe; si no existe, se añade al diccionario.

**Sintaxis:**
```python
diccionario.update({clave: valor})
```
- **clave**: la clave que queremos agregar o modificar.
- **valor**: el valor que queremos asignar a la clave.


````python
def main():
    spacecraft = {"name": "James Webb Space Telescope"} 
    spacecraft.update({"distance": 0.01, "orbit": "Sun"})
    print(create_report(spacecraft))
    

def create_report(spacecraft):
    return f"""
    ========= REPORT =========

    Name: {spacecraft.get('name', "Unknown")}
    Distance: {spacecraft.get('distance', "Unknown")} AU
    Orbit: {spacecraft.get('orbit', "Unknown")}

    =========================
    """

main()
````

## Dictionary `.keys()`

El método `.keys()` devuelve una **vista** con todas las claves de un diccionario.  
Esta vista se comporta como una lista iterable, pero refleja los cambios en el diccionario en tiempo real.

**Sintaxis:**
```python
diccionario.keys()
```
- Devuelve una vista de las claves del diccionario.

````python
distances = {
    "Voyager 1": 163,
    "Voyager 2": 136,
    "Pioneer 10": 80,
    "New Horizons": 58,
    "Pioneer 11": 44
}

def main():
    for name in distances.keys(): # el método .keys() devuelve una vista de las claves del diccionario, es decir, los nombres de las naves espaciales. 
        print(f"{name} is {distances[name]} AU from Earth") # imprime el nombre de la nave espacial (name) y su distancia (distances[name]) en unidades astronómicas (AU).
        # name es una variable que toma el valor de cada clave en cada iteración del bucle, es decir, el nombre de cada nave espacial.

main()
````

## Dictionary `.values()`

El método `.values()` devuelve una **vista** con todos los valores de un diccionario.  
Esta vista es iterable y refleja los cambios en el diccionario en tiempo real.

**Sintaxis:**
```python
diccionario.values()
```
- Devuelve una vista de los valores del diccionario.


````python
distances = {
    "Voyager 1": 163,
    "Voyager 2": 136,
    "Pioneer 10": 80,
    "New Horizons": 58,
    "Pioneer 11": 44
}

def main():
    for distance in distances.values():
        print(f"{distance} AU is {convert(distance)} m") # imprime la distancia en unidades astronómicas (AU) y su conversión a metros (m) utilizando la función convert().

def convert(au):
    return au * 149597870700

main()
````

# Dictionary Methods

Los *Dictionary Methods* son funciones integradas en Python que permiten **interactuar, modificar y consultar** los diccionarios.  
Facilitan tareas como obtener claves, valores, pares, actualizar datos o eliminar elementos.

## Métodos principales

- `.get(clave, valor_predeterminado)` → Devuelve el valor de una clave sin generar error si no existe.
- `.update({clave: valor})` → Modifica o agrega pares clave-valor.
- `.keys()` → Devuelve una vista con todas las claves.
- `.values()` → Devuelve una vista con todos los valores.
- `.items()` → Devuelve una vista con pares clave-valor como tuplas.
- `.pop(clave, valor_predeterminado)` → Elimina una clave y devuelve su valor.
- `.popitem()` → Elimina y devuelve el último par clave-valor.
- `.clear()` → Vacía el diccionario por completo.
- `.copy()` → Devuelve una copia superficial del diccionario.
- `.setdefault(clave, valor)` → Devuelve el valor de una clave; si no existe, la crea con el valor dado.
- `.fromkeys(secuencia, valor)` → Crea un nuevo diccionario con claves de una secuencia y un mismo valor.

**Ejemplo:**
```python
spacecraft = {"name": "Apollo", "year": 1969}

print(spacecraft.get("name"))       # "Apollo"
print(spacecraft.keys())            # dict_keys(['name','year','country'])
print(spacecraft.items())           # dict_items([('name','Apollo'),('year',1969),('country','USA')])
print(spacecraft.values())          # dict_values(['Apollo', 1969, 'USA'])
print(spacecraft.items())           # dict_items([('name', 'Apollo'), ('year', 1969)])
print(spacecraft.clear())           # {}
print(spacecraft.popitem())         # ('country', 'USA')
print(spacecraft.setdefault("mission", "Lunar"))  # "Lunar"
print(spacecraft.fromkeys(["name", "year"], "Unknown"))  # {'name': 'Unknown', 'year': 'Unknown'}
print(spacecraft)                   # {'name': 'Apollo', 'year': 1969}

spacecraft.update({"country": "USA"})
print(spacecraft)                   # {'name': 'Apollo', 'year': 1969, 'country': 'USA'}
spacecraft.pop("year")
print(spacecraft)                   # {'name': 'Apollo', 'country': 'USA'}
backup = spacecraft.copy()           # {} copia del diccionario

```

## Diferencia entre `range(len())` y recorrer directamente la lista

## 🔎 Comparación de los dos enfoques

| Forma | Qué recorre | Ventajas | Desventajas |
|-------|-------------|----------|-------------|
| `for i in range(len(names)):` | Recorre índices (`0, 1, 2, 3`) | - Permite acceder al índice directamente.<br>- Útil si necesitas tanto el valor como la posición.<br>- Puedes manipular el índice (saltar, retroceder, etc.). | - Más verboso.<br>- Menos legible si solo necesitas el valor.<br>- Riesgo de errores si usas mal el índice. |
| `for name in names:` | Recorre directamente los elementos de la lista | - Más limpio y legible.<br>- Evita manejar índices innecesarios.<br>- Es la forma más “pythónica”. | - No tienes acceso al índice (a menos que uses `enumerate`). |

---

## 📌 Cuándo usar cada uno

### ✅ Usar `for name in names`
Cuando solo necesitas trabajar con los valores de la lista.  
Es más claro y evita complejidad innecesaria.

```python
names = ["Mario", "Luigi", "Daisy", "Yoshi"]
for name in names:
    print(write_letter(name, "Princess Peach"))
```

### ✅ Usar `for i in range(len(names))`
Cuando necesitas el índice además del valor.
Ejemplo: imprimir la posición junto al nombre.

```python
names = ["Mario", "Luigi", "Daisy", "Yoshi"]
for i in range(len(names)):
    print(f"{i}: {names[i]}")
```

### ✅ Mejor Alternativa `enumerate`
Si quieres tanto índice como valor, `enumerate` es más elegante y legible.

```python
names = ["Mario", "Luigi", "Daisy", "Yoshi"]
for i, name in enumerate(names):
    print(f"{i}: {write_letter(name, 'Princess Peach')}")
```

## 📊 Ejemplo de salida comparativa

| Código | Salida |
|--------|--------|
| `for name in names` | Mario, Luigi, Daisy, Yoshi |
| `for i in range(len(names))` | 0: Mario, 1: Luigi, 2: Daisy, 3: Yoshi |
| `for i, name in enumerate(names)` | 0: Mario, 1: Luigi, 2: Daisy, 3: Yoshi |

## En resumen
- Usa `for name in names` para iterar directamente sobre los elementos.
- Usa `for i in range(len(names))` si necesitas el índice, pero ten cuidado con la legibilidad.

# List Methods

Los *List Methods* son funciones integradas en Python que permiten **interactuar, modificar y consultar** las listas.  
Facilitan tareas como agregar, eliminar, ordenar o copiar elementos.

## Métodos principales

- `.append(elemento)` → Agrega un elemento al final de la lista.  
- `.extend(iterable)` → Agrega múltiples elementos al final de la lista.  
- `.insert(posición, elemento)` → Inserta un elemento en una posición específica.  
- `.remove(elemento)` → Elimina la primera aparición de un valor.  
- `.pop(posición)` → Elimina y devuelve el elemento en una posición (por defecto el último).  
- `.clear()` → Vacía la lista por completo.  
- `.index(elemento)` → Devuelve el índice de la primera aparición del elemento.  
- `.count(elemento)` → Devuelve cuántas veces aparece un elemento.  
- `.sort(reverse=False)` → Ordena la lista (por defecto ascendente).  
- `.reverse()` → Invierte el orden de los elementos.  
- `.copy()` → Devuelve una copia superficial de la lista.

---

**Ejemplo:**
```python
heroes = ["Mario", "Luigi", "Peach", "Yoshi"]

heroes.append("Toad")
print(heroes)  # ['Mario', 'Luigi', 'Peach', 'Yoshi', 'Toad']

heroes.extend(["Bowser", "Daisy"])
print(heroes)  # ['Mario', 'Luigi', 'Peach', 'Yoshi', 'Toad', 'Bowser', 'Daisy']

heroes.insert(2, "Rosalina")
print(heroes)  # ['Mario', 'Luigi', 'Rosalina', 'Peach', 'Yoshi', 'Toad', 'Bowser', 'Daisy']

heroes.remove("Bowser")
print(heroes)  # ['Mario', 'Luigi', 'Rosalina', 'Peach', 'Yoshi', 'Toad', 'Daisy']

removed = heroes.pop()
print(removed)  # 'Daisy'
print(heroes)   # ['Mario', 'Luigi', 'Rosalina', 'Peach', 'Yoshi', 'Toad']

print(heroes.index("Peach"))  # 3
print(heroes.count("Mario"))  # 1

heroes.sort()
print(heroes)  # ['Luigi', 'Mario', 'Peach', 'Rosalina', 'Toad', 'Yoshi']

heroes.reverse()
print(heroes)  # ['Yoshi', 'Toad', 'Rosalina', 'Peach', 'Mario', 'Luigi']

copy_heroes = heroes.copy()
print(copy_heroes)  # ['Yoshi', 'Toad', 'Rosalina', 'Peach', 'Mario', 'Luigi']

heroes.clear()
print(heroes)  # []


````python
def main():
    history = []

    while True:
        try:
            action = input("Action: ")

            if action == "Undo":
                undone = history.pop()
                print(f"Undone: {undone}")
            elif action == "Restart":
                history.clear()
            else:
                history.append(action)
            
            print(history)
        except KeyboardInterrupt:
            break

main()
````

# String Slicing

El *String Slicing* es una técnica en Python que permite obtener una parte de una cadena de texto (substring) utilizando índices.  
La sintaxis básica es:

```python
cadena[inicio:fin:paso]
```

## 📌 Parámetros

- **inicio** → índice desde donde comienza el corte (incluido).  
- **fin** → índice donde termina el corte (excluido).  
- **paso** → cada cuántos caracteres se avanza (por defecto es `1`).  

---

## 🔎 Ejemplos básicos

```python
text = "abcdef"

print(text[0:4])     # 'abcd' → desde índice 0 hasta 3
print(text[:3])      # 'abc'  → inicio omitido, empieza en 0
print(text[2:])      # 'cdef' → fin omitido, llega hasta el final
print(text[::2])     # 'ace'  → paso de 2, toma cada segundo carácter
print(text[::-1])    # 'fedcba' → paso negativo, invierte la cadena
```

---

## 📊 Tabla de ejemplos comunes

| Expresión | Explicación | Resultado |
|-----------|-------------|-----------|
| `text[0:4]` | Desde índice 0 hasta 3 | `'abcd'` |
| `text[:3]` | Desde inicio hasta índice 2 | `'abc'` |
| `text[2:]` | Desde índice 2 hasta el final | `'cdef'` |
| `text[::2]` | Cada 2 caracteres | `'ace'` |
| `text[::-1]` | Inversión de la cadena | `'fedcba'` |
| `text[5:2:-1]` | Desde índice 5 hasta 3, hacia atrás | `'fed'` |
| `text[4:0:-2]` | Desde índice 4 hasta 1, saltando de 2 en 2 hacia atrás | `'ec'` |

---

## 👉 En resumen
- `inicio` → dónde empieza (incluido).  
- `fin` → dónde termina (excluido).  
- `paso` → salto entre caracteres.  
- Si omites valores, Python usa por defecto: inicio = 0, fin = longitud de la cadena, paso = 1.  
- Con paso negativo, la cadena se recorre en sentido inverso.

````python
def main():
    phone = "+1 617-495-1000"
    print(phone[3::2]) # esta línea de código imprime los caracteres desde el índice 3 hasta el índice 5 (6-1) de la cadena phone, es decir, "617".
    print(phone[8:]) # esta línea de código imprime los caracteres desde el índice 8 hasta el final de la cadena phone, es decir, "495-1000".
    print(phone[-4:]) # esta línea de código imprime los caracteres desde el índice -4 hasta el final de la cadena phone, es decir, "1000".

main()
````

# Tuples

Una `tuple` es una estructura de datos en Python que permite almacenar **múltiples valores** en una sola variable, similar a una lista.  
Sin embargo, a diferencia de las listas, las tuplas son **inmutables**, lo que significa que no se pueden modificar después de su creación.  
Las tuplas se definen utilizando paréntesis `()` y los elementos se separan por comas.

**Ejemplo:**

```python
coordinates = (10.0, 20.0)
print(coordinates[0])  # 10.0
print(coordinates[1])  # 20.0
# Intentar modificar una tupla genera un error
# coordinates[0] = 15.0  # TypeError: 'tuple' object
```

En este ejemplo, se crea una tupla llamada `coordinates` con dos elementos. Se accede a los elementos usando índices, pero si intentamos modificar un elemento de la tupla, se genera un error porque las tuplas son inmutables.

````python
def main():
    coordinates = (42.376, -71.115)
    latitude = coordinates[0]
    longitude = coordinates[1]

    print(f"Latitude: {latitude}, Longitude: {longitude}")
    
main()
````

````python
def main():
    coordinates = (42.376, -71.115)
    latitude, longitude = coordinates # Asigna el primer elemento de la tupla coordinates a la variable latitude y el segundo elemento a la variable longitude.

    print(f"Latitude: {latitude}, Longitude: {longitude}")
    
main()
````

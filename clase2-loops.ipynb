# WHILE

El While loopn es una estructura de control que permite ejecutar un bloque de código repetidamente mientras se cumpla una condición determinada. La sintaxis básica del while loop es la siguiente:

```python
i = 3

while i > 0:
    print("Meow")
    i -= 1
```

```python
i = 3
while i != 0:
    print("Meow")
    i = i - 1
```

```python
i = 1
while i <= 3:
    print("Meow")
    i = i + 1
```

# FOR

El For loop es una forma más común de iterar sobre una secuencia de números, como una lista o un rango de números, en lugar de usar un bucle while con una variable de control. El For loop es más conciso y fácil de leer, y es la forma recomendada de iterar sobre secuencias en Python.

```python
for i in [1, 2, 3]: # esta es una forma de iterar sobre una lista de números con un rango de números manual, pero no es la forma más común de hacerlo en Python.
    print("Meow")
```

```python
for i in range(3): # range(n) genera una secuencia de números desde 0 hasta n-1 y n-1 es el número de iteraciones que se van a realizar
    print("Meow")

for _ in range(3): # la linea _ es un variable de uso común para indicar que no se va a usar el valor de la variable, es decir, que no se va a usar el número de iteraciones.
    print("_Meow")
```

```python
print("Meow" * 3) # esta es una forma de imprimir una cadena de texto un número determinado de veces, en este caso, 3 veces.
print("Meow\n" * 3) #el caracter \n es un salto de línea, por lo que esta línea de código imprimirá "Meow" 3 veces, cada una en una línea diferente.

print("Meow\n" * 3, end="") # el argumento end="" es para evitar que se imprima un salto de línea al final de la última línea, por lo que esta línea de código imprimirá "Meow" 3 veces, cada una en una línea diferente, pero sin un salto de línea al final.
```

```python
while True:
    n = int(input("¿Cuántas veces quieres que se imprima 'Meow'? "))
    if n < 0:
        continue # hace el programa continue con la siguiente iteración del bucle, es decir, vuelve a pedir al usuario que ingrese un número.
    else:
        break # Rompe el bucle, es decir, sale del bucle y continúa con el resto del programa.

for _ in range(n):
    print("Meow")
```

```python
while True:
    n = int(input("¿Cuántas veces quieres que se imprima 'Meow'? "))
    if n > 0:
        break

for _ in range(n):
    print("Meow")
```

# List

Un List es una colección de elementos ordenados y mutables, que se pueden modificar después de su creación. Los elementos de una lista pueden ser de cualquier tipo de datos, como números, cadenas de texto, o incluso otras listas. Las listas se definen utilizando corchetes [] y los elementos se separan por comas.

Ejemplos de como se ve una lista en Python: 

```python
students = ["Hermione", "Harry", "Ron"]

print(students[0]) # imprime el primer elemento de la lista, que es "Hermione"
print(students[1]) # imprime el segundo elemento de la lista, que es "Harry
print(students[2]) # imprime el tercer elemento de la lista, que es "Ron"
```

```python
students = ["Hermione", "Harry", "Ron"]

for student in students:
    print(student) # imprime cada elemento de la lista, es decir, "Hermione", "Harry" y "Ron"
```

```python
students = ["Hermione", "Harry", "Ron"]

for i in range(students): # esta línea de código no es correcta, ya que range() espera un número como argumento, pero students es una lista, por lo que se generará un error.
    print(i + 1, students[i])
```

```python
students = ["Hermione", "Harry", "Ron"]

for i in range(len(students)): # len(students) devuelve el número de elementos en la lista, que es 3 en este caso, por lo que el rango de números será de 0 a 2, es decir, 0, 1 y 2.
    print(i + 1, students[i]) # imprime el número de iteración (i + 1) y el elemento de la lista correspondiente a esa iteración (students[i]), es decir, "1 Hermione", "2 Harry" y "3 Ron".
```

# Dictionary

El Dictionary es una estructura de datos que permite almacenar pares de clave-valor, donde cada clave es única y se utiliza para acceder a su valor correspondiente. En este caso, se puede usar un diccionario para almacenar los nombres de los estudiantes como claves y las casas a las que pertenecen como valores.


> La forma de almacenar en un dictionary es con una clave y un valor representados dentro de un par de llaves ***{}***, donde la clave es el nombre del estudiante y el valor es la casa a la que pertenece, por lo que se puede acceder a la casa de cada estudiante utilizando su nombre como clave.

```python
# No una forma correcta de hacerlo, ya que si se agregan mas datos no da un detalle exacto
students = ["Hermione", "Harry", "Ron", "Draco"]
houses = ["Gryffindor", "Gryffindor", "Gryffindor", "Slytherin"]
```

```python
students = {
    "Hermione": "Gryffindor",
    "Harry": "Gryffindor",
    "Ron": "Gryffindor",
    "Draco": "Slytherin"
}

print(students["Hermione"]) # imprime el valor asociado a la clave "Hermione", que es "Gryffindor", el key es el nombre del estudiante y el value es la casa a la que pertenece.
print(students["Draco"])
```

```python
students = {
    "Hermione": "Gryffindor",
    "Harry": "Gryffindor",
    "Ron": "Gryffindor",
    "Draco": "Slytherin"
}

for student in students:
    print(student, students[student], sep=" - ") # imprime el nombre del estudiante (student) y la casa a la que pertenece (students[student])
```

## List-Dictionary


Se pueden fucionar las listas con dictionarios, que son una estructura de datos que almacena pares de clave-valor, donde cada clave es única y se utiliza para acceder a su valor correspondiente. En este caso, se puede crear un diccionario que asocie cada estudiante con su casa, lo que permite acceder a la información de manera más clara y precisa.

```python
# los Keys son "name", "house" y "patronus", y los values son los datos correspondientes a cada estudiante, es decir, el nombre del estudiante, la casa a la que pertenece y su patronus.
students = [
    {"name": "Hermione", "house": "Gryffindor", "patronus": "Otter"},
    {"name": "Harry", "house": "Gryffindor", "patronus": "Stag"},
    {"name": "Ron", "house": "Gryffindor", "patronus": "Jack Russell Terrier"},
    {"name": "Draco", "house": "Slytherin", "patronus": None} # el None es un valor especial en Python que representa la ausencia de un valor, es decir, que Draco no tiene un patronus.
]

for student in students:
    print(student["name"], student["house"], student["patronus"], sep=" - ") 
    # imprime el nombre del estudiante (student["name"]), la casa a la que pertenece (student["house"]) y su patronus (student["patronus"])
```

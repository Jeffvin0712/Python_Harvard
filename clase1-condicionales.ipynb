# Condicionales

Es la habilidad de preguntar y hacer un proces segun la respuesta. Las condicionales funcionan con formulas matematicas ***<, > <=, >=, ==. !=*** y se comparan con el **IF**



```python
x = int(input("What's x? "))
y = int(input("What's y? "))

if x < y:
    print("x is less than y")
elif x > y:
    print("x is greater than y")
else:
    print("x is equal to y")
```

```python
x = int(input("What's x? "))
y = int(input("What's y? "))

if x < y or x > y:
    print("x is not equal to y")
else:
    print("x is equal to y")
```

```python
x = int(input("What's x? "))
y = int(input("What's y? "))

if x == y:
    print("x is equal to y")
else:
    print("x is not equal to y")
```

```python
score = int(input("What's your score? "))

if score >= 90 and score <= 100:
    print("Your grade is A")
elif score >= 80 and score < 90:
    print("Your grade is B")   
elif score >= 70 and score < 80:
    print("Your grade is C")
elif score >= 60 and score < 70:
    print("Your grade is D")
else:
    print("Your grade is F")
```

# Condicionales con matematicas


```python
x = int(input("What's x? "))

if x % 2 == 0: # Modulo operator returns the remainder of the division of x by 2
    print("x is even") # Even significa que el número es divisible por 2, es decir, no deja residuo al dividirlo por 2
else:
    print("x is odd") # Odd significa que el número no es divisible por 2, es decir, deja un residuo de 1 al dividirlo por 2
```

```python
def main():
    x = int(input("What's x? "))
    if is_even(x):
        print("x is even")
    else:
        print("x is odd")

def is_even(n):
    if n % 2 == 0:
        return True
    else:
        return False

main()
```

# Match method

Es una forma de crear condiciones similares al IF, pero usa una sintaxis diferente

```python
name = input("What's your name? ")

match name:
    case "Harry":
        print("Gryffindor")
    case "Hermione":
        print("Gryffindor")
    case "Ron":
        print("Gryffindor")
    case "Draco":
        print("Slytherin")
    case _:
        print("Who?")
```

```python
name = input("What's your name? ")

match name:
    case "Harry" | "Hermione" | "Ron":
        print("Gryffindor")
    case "Draco":
        print("Slytherin")
    case _:
        print("Who?")
```

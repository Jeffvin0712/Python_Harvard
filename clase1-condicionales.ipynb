# Condicionales
Son estructuras de control que permiten ejecutar diferentes bloques de código según si se cumple o no una condición. Con un `if` evaluamos una expresión booleana y, según el resultado (`verdadero` o `falso`), el programa decide qué camino seguir.

- `a < b`: verdadero si `a` es menor que `b`
- `a > b`: verdadero si `a` es mayor que `b`
- `a <= b`: verdadero si `a` es menor o igual que `b`
- `a >= b`: verdadero si `a` es mayor o igual que `b`
- `a == b`: verdadero si `a` es igual a `b`
- `a != b`: verdadero si `a` no es igual a `b`

Estas comparaciones devuelven valores booleanos (`True` o `False`), y el `if` usa ese resultado para decidir qué código ejecutar.

Ejemplos:
- `x = 5`
- `y = 10`

- `x < y` es `True`
- `x > y` es `False`
- `x <= y` es `True`
- `x >= y` es `False`
- `x == y` es `False`
- `x != y` es `True`

Ejemplos de condicionales en `str`:
- `if nombre == "Alice":`  → compara si la cadena es exactamente `"Alice"`
- `if color != "rojo":`  → comprueba si la cadena no es `"rojo"`

````python
x = int(input("What's x? "))
y = int(input("What's y? "))

if x < y:
    print("x is less than y")
elif x > y:
    print("x is greater than y")
else:
    print("x is equal to y")
````

````python
x = int(input("What's x? "))
y = int(input("What's y? "))

if x < y or x > y:
    print("x is not equal to y")
else:
    print("x is equal to y")
````

````python
x = int(input("What's x? "))
y = int(input("What's y? "))

if x == y:
    print("x is equal to y")
else:
    print("x is not equal to y")
````

# Condicionales con matematicas


````python
x = int(input("What's x? "))

if x % 2 == 0: # Modulo operator returns the remainder of the division of x by 2
    print("x is even") # Even significa que el número es divisible por 2, es decir, no deja residuo al dividirlo por 2
else:
    print("x is odd") # Odd significa que el número no es divisible por 2, es decir, deja un residuo de 1 al dividirlo por 2
````

# Match

`match` es una estructura de control que permite comparar un valor contra varios patrones, similar a un `switch` en otros lenguajes.  
Cada caso se define con `case`, y el primero que coincide se ejecuta.

**Ejemplo:**

```python
x = 2

match x:
    case 1:
        print("Uno")
    case 2:
        print("Dos")
    case _:
        print("Otro")
```
- case 1: se ejecuta si x == 1
- case 2: se ejecuta si x == 2
- case _: es el comodín (default)


````python
name = input("What's your name? ")

match name:
    case "Harry":
        print("Gryffindor")
    case "Hermione":
        print("Gryffindor")
    case "Ron":
        print("Gryffindor")
    case "Draco":
        print("Slytherin")
    case _:
        print("Who?")
````

````python
name = input("What's your name? ")

match name:
    case "Harry" | "Hermione" | "Ron":
        print("Gryffindor")
    case "Draco":
        print("Slytherin")
    case _:
        print("Who?")
````

# Comparación: if/elif/else vs match/case

| Característica        | if/elif/else                          | match/case                          |
|------------------------|---------------------------------------|--------------------------------------|
| Sintaxis              | Usa condiciones booleanas (`==`, `<`) | Compara directamente patrones/valores |
| Flujo                 | Evalúa expresiones una por una        | Busca el primer `case` que coincida   |
| Comodín / Default     | `else`                                | `case _`                             |
| Legibilidad           | Puede volverse largo con muchos `elif`| Más limpio para múltiples opciones    |
| Ejemplo               | `if x == 1:`                         | `match x: case 1:`                   |

# Variables

Una variable es un contenedor que almacena un valor (como un número, texto o dato) que puede cambiar durante la ejecución del programa. Se declara asignándole un nombre y un valor inicial.

```python
nombre = input("What is your name? ")
print(f"Hello, {nombre}!")

#other way to format the string
print("Hello, {}!".format(nombre))
print("Hello, %s!" % nombre)
print("Hello, " + nombre + "!")
```

```python
nombre = input("What is your name? ")

print(f"Hello, {nombre}", end="!") 
#The end parameter is used to specify what to print at the end of the string, 
# by default it is a newline character, but we can change it to a space or anything else

print("Hello,", nombre, sep=" ")
#The sep parameter is used to specify what to print between the arguments,
# by default it is a space, but we can change it to a comma or anything else.
```

```python
nombre = input("What is your name? ")
nombre = nombre.strip() #to remove any leading or trailing whitespace from the input
nombre = nombre.capitalize() #to capitalize the first letter of the input

print(f"Hello, {nombre}!")
```

```python
nombre = input("What is your name? ")
nombre = nombre.strip() #to remove any leading or trailing whitespace from the input
nombre = nombre.title() #to capitalize the first letter of each word in the input

print(f"Hello, {nombre}!")
```

```python
nombre = input("What is your name? ").strip().title()

print(f"Hello, {nombre}!")
```

```python
nombre = input("What is your name? ").strip().title()

# Split user's name into first and last name
first_name, last_name = nombre.split(" ")

print(f"Hello, {first_name} {last_name}!")
```

### INT Variables

Existen multiples maneras de como poder representar los numeros en las variables. Asignandolo desde la tercera variable, en el inicio o todo en una sola linea (no tan recomedable)

```python
x = input("Enter a number: ")
y = input("Enter another number: ")

z = int(x) + int(y)
print(z)
```

### Float Variables

Permite la matematica con numeros decimales, este funciona con el predeterminado de la cantidad x de decimales o se puede redondear o limitar con ROUND o la funcion ***z:,.2f***

- ***round(float(variable), n)***
- ***round(variable, n)***

Para darle formato de separar los miles, se puede usar la expresion ***print(f"{z:,}")***


```python
x = float(input("Enter a number: "))
y = float(input("Enter another number: "))

z = round(x + y, 2)

print(f"{z:,}")
```

```python
x = float(input("Enter a number: "))
y = float(input("Enter another number: "))

z = x + y

print(f"{z:,.2f}")
```

# Functions

Son programas que se crean para que realicen una acción determinada.
Las funciones se determinan con el **def name():** y despues los argumentos que va a hacer el proceso.

En condiciones normales, Python lee los procesos de arriba para abajo, por lo que es necesario que las funciones esten en la parte de arriba para que cuando se llamen, no genere problemas.

- Las funcionan por medio de ***scope***, que quiere decir que solamente las variables existen dentro de las funciones
- Para que la funcion me de información como resultado de su proceso, se tiene que usar ***return***

```python
def hello(to="World"): # La función hello tiene un parámetro llamado to, que tiene un valor por defecto "World". 
    print(f"Hello, {to}!") # Si no se le pasa un argumento al llamar a la función, el valor por defecto se usará

name = input("What is your name? ")
hello(name) # el name es un argumento, es un valor que se pasa a la función cuando se llama a la función
hello() # Llama a la función sin argumentos, usando el valor por defecto "World"
```

```python
def main():
    name = input("What is your name? ")
    hello(name)

def hello(to="World"):
    print(f"Hello, {to}!")

main()
```

```python
def main():
    x = int(input("What's x: "))
    print("x squared is", square(x))

def square(n):
    return pow(n, 2) # pow() es una función incorporada en Python que toma dos argumentos y devuelve el primer argumento elevado a la potencia del segundo argumento.

main()
```

## Return Values (def)

```python
def area(length, width):
    print(str(length * width) + " square feet")

def main():
    house = area(5, 10) # Llama a la función area con los argumentos 5 y 10, lo que imprimirá "50 square feet".
    print(house) # Esto imprimirá None, ya que la función area no devuelve ningún valor, sino que solo imprime el resultado.

main() # Main funciona llamando a la función area con los argumentos 5 y 10, lo que imprimirá "50 square feet".

```

```python
def area(length, width):
    return length * width # La función area ahora devuelve el área en lugar de imprimirla, lo que permite usar el valor devuelto en otras partes del programa.

def main():
    house = area(5, 10)
    yard = area(50, 50)
    print(f"{house} square feet")
    print(f"{yard} square feet")

main() 
```

```python
def area(length, width):
    print(str(length * width) + " square feet")
    return length * width # La función area ahora devuelve el área en lugar de imprimirla, lo que permite usar el valor devuelto en otras partes del programa.

def main():
    house = area(5, 10)
    yard = area(50, 50)
    total = house + yard
    print(f"{total} square feet")

main() 
```

# Methods

## String Methods

Método | Descripción | Ejemplo
--- | --- | ---
capitalize | Convierte el primer carácter en mayúscula y el resto en minúsculas. | `"hello".capitalize()` → `"Hello"`
casefold | Devuelve una copia en minúsculas para comparaciones sin distinción de mayúsculas. | `"Hello".casefold()` → `"hello"`
center | Centra la cadena en un ancho dado rellenando con espacios u otro carácter. | `"hi".center(10)` → `"    hi    "`
count | Cuenta cuántas veces aparece un subcadena. | `"banana".count("a")` → `3`
encode | Codifica la cadena en bytes usando una codificación dada. | `"hello".encode("utf-8")` → `b'hello'`
endswith | Comprueba si la cadena termina con una subcadena dada. | `"hello".endswith("lo")` → `True`
expandtabs | Reemplaza tabulaciones por espacios. | `"h\te\tl".expandtabs(4)` → `"h   e   l"`
find | Busca una subcadena y devuelve el índice de la primera aparición o -1. | `"hello".find("e")` → `1`
format | Formatea la cadena usando los valores provistos. | `"Hi {}".format("world")` → `"Hi world"`
format_map | Formatea la cadena usando un diccionario o mapeo. | `"Hi {name}".format_map({"name": "world"})` → `"Hi world"`
index | Busca una subcadena y devuelve el índice de la primera aparición; falla si no la encuentra. | `"hello".index("e")` → `1`
isalnum | Comprueba si todos los caracteres son alfanuméricos. | `"abc123".isalnum()` → `True`
isalpha | Comprueba si todos los caracteres son letras. | `"abc".isalpha()` → `True`
isascii | Comprueba si todos los caracteres son ASCII. | `"hello".isascii()` → `True`
isdecimal | Comprueba si todos los caracteres son decimales. | `"123".isdecimal()` → `True`
isdigit | Comprueba si todos los caracteres son dígitos. | `"123".isdigit()` → `True`
isidentifier | Comprueba si la cadena es un identificador válido de Python. | `"var1".isidentifier()` → `True`
islower | Comprueba si todos los caracteres alfabéticos están en minúsculas. | `"hello".islower()` → `True`
isnumeric | Comprueba si todos los caracteres son numéricos. | `"123".isnumeric()` → `True`
isprintable | Comprueba si todos los caracteres son imprimibles. | `"hello".isprintable()` → `True`
isspace | Comprueba si todos los caracteres son espacios en blanco. | `"   ".isspace()` → `True`
istitle | Comprueba si la cadena está en formato título. | `"Hello World".istitle()` → `True`
isupper | Comprueba si todos los caracteres alfabéticos están en mayúsculas. | `"HELLO".isupper()` → `True`
join | Une iterable de cadenas usando la cadena como separador. | `" ".join(["a", "b"])` → `"a b"`
ljust | Alinea la cadena a la izquierda en un ancho dado. | `"hi".ljust(10)` → `"hi        "`
lower | Convierte todos los caracteres a minúsculas. | `"HELLO".lower()` → `"hello"`
lstrip | Elimina espacios u otros caracteres al principio. | `"  hello".lstrip()` → `"hello"`
maketrans | Crea una tabla de traducción para `translate`. | `str.maketrans("a", "b")` → tabla para reemplazar 'a' por 'b'
partition | Divide la cadena en tres partes usando un separador. | `"a b c".partition(" ")` → `("a", " ", "b c")`
replace | Reemplaza ocurrencias de una subcadena por otra. | `"hello".replace("l", "x")` → `"hexxo"`
rfind | Busca una subcadena empezando desde la derecha; devuelve -1 si no la encuentra. | `"hello".rfind("l")` → `3`
rindex | Busca una subcadena empezando desde la derecha; falla si no la encuentra. | `"hello".rindex("l")` → `3`
rjust | Alinea la cadena a la derecha en un ancho dado. | `"hi".rjust(10)` → `"        hi"`
rpartition | Divide la cadena en tres partes por la última ocurrencia del separador. | `"a b c".rpartition(" ")` → `("a b", " ", "c")`
rsplit | Divide la cadena en una lista desde la derecha. | `"a b c".rsplit(" ")` → `["a", "b", "c"]`
rstrip | Elimina espacios u otros caracteres al final. | `"hello   ".rstrip()` → `"hello"`
split | Divide la cadena en una lista usando un separador. | `"a b c".split(" ")` → `["a", "b", "c"]`
splitlines | Divide la cadena en una lista por líneas. | `"a\nb".splitlines()` → `["a", "b"]`
startswith | Comprueba si la cadena empieza con una subcadena dada. | `"hello".startswith("he")` → `True`
strip | Elimina espacios u otros caracteres al principio y al final. | `"  hello  ".strip()` → `"hello"`
swapcase | Invierte mayúsculas y minúsculas. | `"Hello".swapcase()` → `"hELLO"`
title | Convierte la cadena a formato título. | `"hello world".title()` → `"Hello World"`
translate | Reemplaza caracteres según una tabla de traducción. | `"hello".translate(str.maketrans("l", "x"))` → `"hexxo"`
upper | Convierte todos los caracteres a mayúsculas. | `"hello".upper()` → `"HELLO"`
zfill | Rellena la cadena con ceros a la izquierda hasta un ancho dado. | `"42".zfill(5)` → `"00042"`
removeprefix | Elimina un prefijo si la cadena lo tiene. | `"hello world".removeprefix("hello ")` → `"world"`
removesuffix | Elimina un sufijo si la cadena lo tiene. | `"hello world".removesuffix(" world")` → `"hello"`

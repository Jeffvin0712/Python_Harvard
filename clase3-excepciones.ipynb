# Errores

Los errores en Python pueden ser de distintos tipos. Aquí se presenta una lista ordenada de los errores más comunes:

- **SyntaxError**: Ocurre cuando el código no sigue las reglas de sintaxis del lenguaje de programación. En este caso, falta una comilla al final de la cadena de texto.
- **NameError**: Ocurre cuando se intenta usar una **variable que no ha sido definida** previamente.
- **TypeError**: Ocurre cuando se intenta realizar una operación o función en un **tipo de dato inapropiado**.
- **ValueError**: Ocurre cuando se pasa un **valor no válido** a una función.
- **IndexError**: Ocurre cuando se intenta acceder a un **índice fuera del rango** de una lista o secuencia.
- **KeyError**: Ocurre cuando se intenta acceder a una **clave inexistente** en un diccionario.
- **AttributeError**: Ocurre cuando se intenta acceder a un **atributo o método que no existe** en un objeto.
- **ZeroDivisionError**: Ocurre cuando se intenta **dividir por cero**.
- **FileNotFoundError**: Ocurre cuando se intenta abrir un **archivo que no existe**.
- **ImportError**: Ocurre cuando se intenta importar un **módulo no accesible**.
- **MemoryError**: Ocurre cuando el programa **se queda sin memoria disponible**.
- **RecursionError**: Ocurre cuando una función se llama a **sí misma demasiadas veces**.
- **IOError**: Ocurre cuando se produce un **error de entrada/salida** al leer o escribir archivos.

# Exceptions

Las excepciones en Python son eventos que ocurren durante la ejecución de un programa y que interrumpen el flujo normal del mismo. Cuando Python encuentra un error, genera una excepción que puede ser capturada y manejada para evitar que el programa se detenga abruptamente.

## Características principales:

- **Son objetos**: Las excepciones son instancias de clases que heredan de la clase `BaseException`.
- **Interrumpen la ejecución**: Cuando se produce una excepción no manejada, el programa se detiene y muestra un mensaje de error.
- **Pueden ser capturadas**: Utilizando bloques `try-except`, podemos atrapar excepciones y definir cómo responder ante ellas.
- **Permiten programación defensiva**: Con el manejo de excepciones, podemos prever errores potenciales y crear código más robusto.

## Diferencia entre errores y excepciones:

- **Errores**: Son problemas graves que ocurren durante la ejecución (como en la celda 1 vemos un `SyntaxError`).
- **Excepciones**: Son errores que pueden ser anticipados y manejados durante la ejecución del programa (como `ValueError` o `IndexError`).

## Ejemplo de manejo de excepciones:

El código en la celda 2 que solicita un número al usuario podría generar una `ValueError` si el usuario ingresa texto en lugar de un número. Con excepciones, podríamos manejar este caso de forma controlada en lugar de permitir que el programa falle.





```python
try:
    x = int(input("What's x? "))
    print(f"x is {x}")
except ValueError:
    print("x is not an integer")
```

```python
# Este falla porque x es validado por el except y no se asigna un valor a x, por lo que el print falla porque x no existe
try:
    x = int(input("What's x? "))
except ValueError:
    print("x is not an integer")

print(f"x is {x}") # Esto funcionaria solo si el numero ingresado es un entero, de lo contrario x no se asigna y el print falla
```

```python
try:
    x = int(input("What's x? "))
except ValueError:
    print("x is not an integer")
else:
    print(f"x is {x}")
```

```python
while True:
    try:
        x = int(input("What's x? "))
    except ValueError:
        print("x is not an integer")
    else:
        break
    
print(f"x is {x}")
```

```python
while True:
    try:
        x = int(input("What's x? "))
        break
    except ValueError:
        print("x is not an integer")
    
print(f"x is {x}")
```

```python
def main():
    x = get_int()
    print(f"x is {x}")

def get_int():
    while True:
        try:
            x = int(input("What's x? "))
            break
        except ValueError:
            print("x is not an integer")
    return x

main()
```

# Pass

El `pass` en un bloque `try-except` es una declaración que no realiza ninguna acción. Se utiliza para ignorar la excepción capturada, permitiendo que el programa continúe su ejecución sin manejar el error de manera específica. Es útil cuando no se requiere una respuesta particular ante la excepción.

```python
def main():
    x = get_int()
    print(f"x is {x}")

def get_int():
    while True:
        try:
            return int(input("What's x? "))
        except ValueError:
            pass # Esto hace que el programa no haga nada si el usuario ingresa un valor no entero, simplemente vuelve a pedir el input

main()
```

```python
def main():
    x = get_int("What's x? ")
    print(f"x is {x}")

def get_int(prompt):
    while True:
        try:
            return int(input(prompt))
        except ValueError:
            pass 

main()
```

# Raise

En un bloque `try-except`, `raise` se usa para volver a lanzar una excepción o para lanzar una excepción nueva desde dentro del `except`. Esto permite:

- propagar el error hacia el código que llamó a la función,
- o transformar un error en otro más adecuado al contexto.

Por ejemplo:
- dentro de `except`, `raise` sin argumentos vuelve a lanzar la excepción capturada,
- `raise ValueError("mensaje")` lanza una nueva excepción explícita.

Así se puede manejar un error parcialmente y luego dejar que el programa lo trate en otro nivel.

```python
def main():
    pace = get_page(miles=26.2, minutes=0)
    print(f"You need to run each mile in {round(pace, 2)} minutes")

def get_page(miles, minutes):
    if not minutes > 0:
        raise Exception()
    return minutes / miles

main()
```

# Debugging

- Identificar el problema observando el comportamiento del programa y el mensaje de error.
- Leer el código relacionado con la sección que falla.
- Usar `print()` o las herramientas del depurador para inspeccionar valores de variables en puntos clave.
- Ejecutar paso a paso el código para ver en qué momento cambia el estado inesperadamente.
- Verificar entradas, tipos de datos y condiciones de los `if`, bucles y funciones.
- Corregir el error y volver a probar hasta que el programa funcione como se espera.

```python
def main():
    height = int(input("Height: "))
    pyramid(height)

def pyramid(n):
    for i in range(n):
        print(i, end=" ")
        print("#" * i)

if __name__ == "__main__":
    main()
```

## Breakpoint

Un breakpoint en VS Code es un punto de interrupción que colocas en una línea de código para que el depurador detenga la ejecución en ese sitio. Al llegar a un breakpoint, puedes:

- ver el valor de variables,
- inspeccionar la pila de llamadas,
- ejecutar paso a paso (`step over`, `step into`, `continue`),
- y detectar el estado del programa en ese momento.

En VS Code se activa haciendo clic en el margen izquierdo junto al número de línea o usando `F9`, y se ve como un punto rojo. Esto ayuda a encontrar errores y entender cómo fluye el código sin modificarlo con `print()`.

```python
def main():
    height = int(input("Height: "))
    pyramid(height)

def pyramid(n):
    for i in range(n):
        print("#" * (i + 1)) # al hacer el i + 1, el numero de # empieza en 1 y no en 0, lo que hace que la piramide tenga una altura de n y no de n - 1

if __name__ == "__main__":
    main()
```

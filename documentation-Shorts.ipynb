## Funciones

* __def xxx()__: is a function that allows to creaste multiples processes or sequences needed for the program. Can storage _variables_ or ***actions***.

We can use 2 functions to complete some actions for the other function like the example below

>>Into the functions is needed to do a return to allow the second function to have the information

```python
def area(length, width):
    print(str(length * width) + " square units")
    return length * width

def main():
    house = area(20, 20)
    yard = area(50, 50)
    total_area = house + yard
    print(f"The total area of the house and yard is: {total_area} square units.")

main()
```

## Variables Globales

To update a variable that is outside of the function is needed to call it with ***global*** + name of the variable

```python
emoticon = "v.v"

def main():
    global emoticon
    say("Is anyone there?")
    emoticon = ":D"
    say("I am here!")


def say(phrase):
    print(phrase + " " + emoticon)

main()
```

## Condicionales

With this you can validate in form of boolean expresion. All of this based on the user's input.

```python
def main():
    difficulty = input("Difficult or Casual? ").capitalize()
    if difficulty not in ["Difficult", "Casual"]:
        print("Invalid input. Please choose 'Difficult' or 'Casual' for difficulty.")
        return

    players = input("Multiplayer or Singleplayer? ").capitalize()
    if players not in ["Multiplayer", "Singleplayer"]:
        print("Invalid input. Please choose 'Multiplayer' or 'Singleplayer' for players.")
        return

    if difficulty == "Difficult" and players == "Multiplayer":
        recommend("Poker")
    elif difficulty == "Difficult" and players == "Singleplayer":
        recommend("Chess")
    elif difficulty == "Casual" and players == "Multiplayer":
        recommend("Uno")
    else:
        recommend("Solitaire")


def recommend(game):
    print("You might like " + game)

main()
```

### Métodos de Diccionarios

When we call the variable we can access to the key or value. The keys of the dictionary are the words, and the values are the points for each word

```python
WORDS = {"PAIR": 4, "HAIR": 4, "CHAIR": 5, "GRAPHIC": 7} #The keys of the dictionary are the words, and the values are the points for each word


def main():
    print("Welcome to Spelling Bee!")
    print("Your letters are: A I P C R H G")

    while len(WORDS) > 0:
        print(f"{len(WORDS)} words left!")
        guess = input("Guess a word: ").upper()

        # TODO: Check if guess in dictionary
        if guess == "GRAPHIC":
            WORDS.clear()
            print("You've won!")

        if guess in WORDS.keys(): #WORDS.keys() returns a list of the keys in the dictionary, which are the words in this case
            points = WORDS.pop(guess) #Remove the word from the dictionary and return the points
            print(f"Good jon! You scored {points} points.")


    print("That's the game!")

main()
```

# Métodos de Diccionarios

| **Método** | **Descripción (ES)** | **Ejemplo** |
|------------|----------------------|-------------|
| `clear()` | Elimina todos los elementos del diccionario | `d = {'a': 1}; d.clear()` → `{}` |
| `copy()` | Devuelve una copia superficial del diccionario | `d = {'a': 1}; d2 = d.copy()` → `{'a': 1}` |
| `fromkeys(iterable, value=None)` | Crea un nuevo diccionario con claves de un iterable y valor opcional | `dict.fromkeys(['a', 'b'])` → `{'a': None, 'b': None}` |
| `get(key, default=None)` | Devuelve el valor de la clave, o default si no existe | `d = {'a': 1}; d.get('a')` → `1`; `d.get('b', 0)` → `0` |
| `items()` | Devuelve una vista de pares clave-valor | `d = {'a': 1}; list(d.items())` → `[('a', 1)]` |
| `keys()` | Devuelve una vista de las claves | `d = {'a': 1}; list(d.keys())` → `['a']` |
| `pop(key, default)` | Elimina y devuelve el valor de la clave, o default si no existe | `d = {'a': 1}; d.pop('a')` → `1`; `d` → `{}` |
| `popitem()` | Elimina y devuelve el último par clave-valor | `d = {'a': 1, 'b': 2}; d.popitem()` → `('b', 2)` |
| `setdefault(key, default=None)` | Devuelve el valor si existe, sino lo establece y devuelve default | `d = {'a': 1}; d.setdefault('b', 2)` → `2`; `d` → `{'a': 1, 'b': 2}` |
| `update(other)` | Actualiza el diccionario con pares de otro diccionario o iterable | `d = {'a': 1}; d.update({'b': 2})` → `{'a': 1, 'b': 2}` |
| `values()` | Devuelve una vista de los valores | `d = {'a': 1}; list(d.values())` → `[1]` |

### List methods

A list method is a function that is built into the list data type in Python. It allows you to perform specific operations on lists, such as adding, removing, or modifying elements. Some common list methods include **append(), extend(), remove(), insert(), reverse(), and sort()**. These methods provide a convenient way to manipulate lists without having to write custom code for each operation.



```python
def main():
    history = []

    while True:
        action = input("Action: ")

        if action == "Undo":
            undone = history.pop() # take off the last element
            print(f"Undone: {undone}")
        elif action == "Restart":
            history.clear()
        else:
            history.append(action)

        print(history)

main()
```

# Métodos de Listas

| **Método** | **Descripción (ES)** | **Ejemplo** |
|------------|----------------------|-------------|
| `append(item)` | Agrega un elemento al final de la lista | `lst = [1, 2]; lst.append(3)` → `[1, 2, 3]` |
| `clear()` | Elimina todos los elementos de la lista | `lst = [1, 2]; lst.clear()` → `[]` |
| `copy()` | Devuelve una copia superficial de la lista | `lst = [1, 2]; lst2 = lst.copy()` → `[1, 2]` |
| `count(item)` | Cuenta las ocurrencias de un elemento | `lst = [1, 2, 2]; lst.count(2)` → `2` |
| `extend(iterable)` | Extiende la lista con elementos de un iterable | `lst = [1, 2]; lst.extend([3, 4])` → `[1, 2, 3, 4]` |
| `index(item, start, end)` | Devuelve el índice de la primera ocurrencia | `lst = [1, 2, 3]; lst.index(2)` → `1` |
| `insert(index, item)` | Inserta un elemento en la posición especificada | `lst = [1, 2]; lst.insert(1, 3)` → `[1, 3, 2]` |
| `pop(index=-1)` | Elimina y devuelve el elemento en la posición (último por defecto) | `lst = [1, 2, 3]; lst.pop()` → `3`; `lst` → `[1, 2]` |
| `remove(item)` | Elimina la primera ocurrencia del elemento | `lst = [1, 2, 3]; lst.remove(2)` → `[1, 3]` |
| `reverse()` | Invierte el orden de los elementos | `lst = [1, 2, 3]; lst.reverse()` → `[3, 2, 1]` |
| `sort(key=None, reverse=False)` | Ordena los elementos de la lista | `lst = [3, 1, 2]; lst.sort()` → `[1, 2, 3]` |

```python
# Example to validate a vanity license plate, which must start with at least two letters, followed by a combination of letters and numbers, 
# but the numbers cannot start with 0 and must be at the end of the plate. The plate must be between 2 and 6 characters long.

def main():
    plate = input("Plate: ")
    if is_valid(plate):
        print("Valid")
    else:
        print("Invalid")


def is_valid(s):
    if not (2 <= len(s) <= 6): #The length of the plate must be between 2 and 6 characters
        return False

    if not s[0].isalpha() or not s[1].isalpha(): #The first two characters must be letters
        return False

    if not s.isalnum(): #The plate must be alphanumeric, meaning it can only contain letters and numbers
        return False

    for i in range(len(s)): # Loop through the characters in the plate to find the first number
        if s[i].isdigit():
            if s[i] == "0":
                return False

            if not s[i:].isdigit(): # The numbers must be at the end of the plate, so if there are any letters after the first number, it is invalid
                return False
            break

    return True


main()
```
